## 1. Setup Environment

```python
from Preprocessing.selfsupervised_data_praparation import plot_samples, get_dataset_list, split_list, CustomDataGen, show_random_images, plot_autoencoder_predictions
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import tensorflow as tf
import matplotlib.pyplot as plt
```

### Constants

```python
# Path to Dataset
path = '../Dataset_10_best_cities/'
model_name = 'autoencoder' #'variational_autoencoder'
latent_dim = 1024

# Dataloader Args:
batch_size = 16
target_size = (224, 224, 3)
```

## 2. Read the Dataset

```python
#plot_samples(path)
```

```python
image_list = get_dataset_list(path, show_dirs=True, head=10)
```

## 3. Dataset Preparation

### 3.1. Train-Test-Split

```python
x_train, x_test = split_list(image_list)
```

### 3.2. Augmentation

```python
# Augmentation for train
train_datagen = ImageDataGenerator(
    rotation_range=10,
    zoom_range=0.5,
    width_shift_range=0.1,
    height_shift_range=0.1,
    horizontal_flip=True,
    vertical_flip=True)
```

### 3.3. Create a custom Dataloader

```python
traingen = CustomDataGen(x_train,
                         batch_size=batch_size, 
                         input_size=target_size, 
                         augmentation=train_datagen, 
                         shuffle = True)
```

```python
testgen = CustomDataGen(x_test,
                         batch_size=batch_size, 
                         input_size=target_size, 
                         augmentation=None)
```

```python
show_random_images(x_test, augmentation=train_datagen, target_size=target_size)
```

## 3. Model

### 3.3. Create the Model

```python
from Models import Autoencoder
from Models import Variational_Autoencoder
```

```python
tf.random.set_seed(0)

if model_name == 'variational_autoencoder':
    model = Variational_Autoencoder.get_Variational_Autoencoder(target_size = target_size, latent_dim = latent_dim, lr=0.0001)
elif model_name == 'autoencoder':
    model = Autoencoder.get_Autoencoder(target_size = target_size, latent_dim = latent_dim)

model.summary()
```

### 3.2. Train model

```python
from tensorflow.keras.callbacks import EarlyStopping

# EarlyStopping:
monitor = EarlyStopping(monitor='val_loss', min_delta=1e-4, patience=10, 
        verbose=1, mode='auto', restore_best_weights=True)
```

```python
# fit the model
history = model.fit(traingen,
                validation_data=testgen,
                epochs=100,
                callbacks=[monitor]
)
```

```python
# plot history
plt.plot(history.history['loss'], label='train')
plt.plot(history.history['val_loss'], label='test')
plt.legend()
plt.show()
```

### 3.3. Test the Model

```python
""" Predictions """
plot_autoencoder_predictions(model, testgen, samples=6)
```

```python
model.evaluate(testgen)
```

### Save the model

```python
# Save the model
model.save(f'Weights/{model_name}_{target_size[0]}_{latent_dim}.h5')
```
